This is data from Swiss Federal Statistical Office https://www.pxweb.bfs.admin.ch/Default.aspx?px_language=en Interactive data  tables can be built in https://www.pxweb.bfs.admin.ch/Selection.aspx?px_language=en&px_db=px-x-0102010000_104&px_tableid=px-x-0102010000_104\px-x-0102010000_104.px&px_type=PX

The dataset below has been exported as "Tab limited with heading" with "Table - layout 2" option.

The entries are not separated by commas but by tabulators, and the headings are in the second line of the file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

countries = pd.read_table('data/px-x-0102010000_104.csv', sep='\t', encoding = "ISO-8859-1")
countries.head()
```

Here we had as a first row an additional title visible in the far right so we take the header from row 2 (header = 1)

```python
countries = pd.read_table('data/px-x-0102010000_104.csv', header=1, sep='\t', encoding = "ISO-8859-1")
countries.head()
```

Take the values starting from Switzerland (entry number 5):

```python
labels = list(countries)[5:]
print(labels)
```

```python
entries = list(countries.ix[0])[5:]
print(entries)
```

```python

#this is needed to get the plot right at the first execution
plt.rcdefaults()

#setting the size of the plot big enough for the text to be readable
fig, ax = plt.subplots(figsize=(10,40))

#this makes a range of values for y axis, which will then be replaced by the country names
y_pos = np.arange(len(labels))
ax.barh(y_pos, entries,
        color='green')

ax.set_yticks(y_pos)
ax.set_yticklabels(labels)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of persons per nationality')
ax.set_title('Nationalities of the permanent resident population in Meyrin')
ax.set_xscale('log')
plt.grid()


plt.show()
```
