Fysiikan ylioppilastutkinnnot (ilmoittautuneet). NB: sisältää myös kokelaat, jotka uusivat aiemmin suoritettuja kokeita.

Lähde: 
- https://www.ylioppilastutkinto.fi/ext/stat/FS2018A2009T2010.pdf
- https://www.stat.fi/til/lop/2017/lop_2017_2018-06-13_tau_003_fi.html


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vuosi = np.array([2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
```

```python
 kaikki = np.array([32643, 32324, 32413, 31595, 31693, 31620, 30617, 30531, 30189])
female = np.array([18835, 18752, 18571, 18244, 18181, 18074, 17589, 17834, 17518])
male = kaikki - female
```

```python
plt.scatter(vuosi, kaikki, label='Ylioppilastutkinnot kaikki')
plt.scatter(vuosi, female, label='Ylioppilastutkinnot tytöt')
plt.scatter(vuosi, male, label='Ylioppilastutkinnot pojat')
plt.legend(loc='best')
plt.ylim(ymin=0)
plt.show()
```

```python
fysiikka = np.array([5914, 5947, 6106, 6392, 6313, 6554, 6406, 6669, 6712])
fysmalepro = np.array([78.8, 76.5, 76.5, 73.0, 74.8, 72.8, 70.0, 66.9, 66.1])
fysfemalepro = np.array([21.2, 23.5, 23.5, 27.0, 25.2, 27.2, 30.0, 33.1, 33.9])
fysmale = fysiikka*fysmalepro/100
fysfemale = fysiikka*fysfemalepro/100
```

```python
ratio = 100*fysiikka/kaikki
ratiomale = 100*fysmale/male
ratiofemale = 100*fysfemale/female
```

```python
plt.scatter(vuosi, ratio, label='Fysiikan kirjoittajat kaikista (%)')
plt.scatter(vuosi, ratiomale, label='Fysiikan kirjoittajat pojista (%)')
plt.scatter(vuosi, ratiofemale, label='Fysiikan kirjoittajat tytöistä (%)')

plt.legend(loc='upper left')
plt.ylim(ymin=0, ymax=45)

plt.show()

plt.scatter(vuosi, fysiikka, label='Fysiikan kirjoittajat')
plt.scatter(vuosi, fysmale, label='Fysiikan kirjoittajat pojat')
plt.scatter(vuosi, fysfemale, label='Fysiikan kirjoittajat tytöt')

plt.legend(loc='best')
plt.ylim(ymin=0)

plt.show()
```

```python
plt.scatter(vuosi, fysiikka/fysiikka[0], label='Fysiikan kirjoittajat kaikista, muutos vuodesta 2009')
plt.scatter(vuosi, fysmale/fysmale[0], label='Fysiikan kirjoittajat pojista, muutos vuodesta 2009')
plt.scatter(vuosi, fysfemale/fysfemale[0], label='Fysiikan kirjoittajat tytöistä, muutos vuodesta 2009 ')
plt.ylim(ymin=0, ymax=3)

plt.legend(loc='upper left')

plt.show()
```
