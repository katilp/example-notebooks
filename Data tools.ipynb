# Considerations for tools to propose in education

This is an example notebook to illustrate considerations to be taken into account for data tools that can be proposed at different stages in education.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Different groups: junior high (lower secondary), high school (upper secondary), university basic courses, university research education

```python
group = np.array(['12-15y', '15-18y', 'Univ (basic)', 'Univ (research)'])
```

These are very rough estimates for Finland

*Audience* in different groups, sources:
- Number of students in lower secondary: https://www.stat.fi/til/pop/2017/pop_2017_2017-11-14_tau_001_fi.html (take 25% of last year's students)
- Students taking physics matriculation exam yearly:  https://www.ylioppilastutkinto.fi/ext/stat/FS2018A2009T2010.pdf   
- Numbers for basic/upper level university students in sciences: https://www.stat.fi/til/yop/2016/yop_2016_2017-05-10_tau_001_fi.html (take 50% of yearly natural science students for bachelor and MSc, respectively)

*Time*: estimate of the time available for an exercise (hours)

*Tools*: estimate of the complexity of tools to be proposed for each group (relative value, max = 20)

```python
audience = np.array([15000, 6000, 1000,200])
time = np.array(['1.5','1.5','5','20'])
tools = np.array([1,2,4,20])
```

(Python help: two y axes in one plot https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib)

```python
x = np.arange(4)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x, audience, color='b')
ax2.plot(x, time, 'r-', label = 'Time available')
ax2.plot(x, tools, 'r--', label = 'Tools complexity')

plt.xticks(x, group)


ax1.set_xlabel('Group')
ax1.set_ylabel('Audience', color='b')
ax2.set_ylabel('Time available(h)/Tools complexity(rel)', color='r')
plt.legend(loc='upper right')


plt.show()
```
