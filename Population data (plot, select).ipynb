This is data from https://catalog.data.gov/dataset (csv from https://catalog.data.gov/dataset/2010-census-populations-by-zip-code)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data/2010_Census_Populations_by_Zip_Code.csv')
```

```python
dataset.head()
```

```python
hh_size = dataset['Average Household Size']
```

```python
plt.hist(hh_size, bins=20)
plt.show()
```

```python
med_age = dataset['Median Age']
plt.hist(med_age, bins=40)
plt.show()
```

```python
% matplotlib inline
female_to_male = dataset['Total Females']/dataset['Total Males']
print(female_to_male)
```

```python
#Ups, there is at least one inf, so make a selection to exclude that
newdata = dataset[dataset['Total Males']>0]
female_to_male = newdata['Total Females']/newdata['Total Males']
plt.hist(female_to_male, bins=60)
plt.show()
```

```python
female_to_male.mean()
```

```python
female_to_male.std()
```
